```python
def Tokenize(Text):
    Text = re.sub(r'[^\w\s]','',Text)
    Text = Text.lower()
    Text_liste = Text.split(" ")
    return Text_liste
```

```python
df["Discour"] = df["Discour"].apply(Tokenize)
```

```python
df["Discour"]
```

```python
def Term_frequency(My_list):
    D = {}
    for elem in My_list :
        if elem not in D.keys():
            cnt = My_list.count(elem)
            D[elem] = cnt
    return D
```

```python
df["TF"] = df["Discour"].apply(Term_frequency)
```

```python
df["TF"]
```

```python
df
```

```python
def Term_frequency_Documents(column):
    D = {}
    L = []
    for item in df["TF"].iteritems():
        L += list(item[1].keys())
    D = Term_frequency(L)
    return D
```

```python
TFD = Term_frequency_Documents(df["TF"])
```

```python
TFD
```

```python
def Inverse_Document_Frequency(dic):
    D = {}
    for item in dic.items():
        D[item[0]] = math.log(205/item[1])
    return D
```

```python
IDF = Inverse_Document_Frequency(TFD)
```

```python
IDF
```

```python
def TF_IDF(dic):
    global IDF
    D = {key: dic[key]*IDF[key] for key in dic}
    return D
```

```python
df["TF-IDF"] = df["TF"].apply(TF_IDF)
```

```python
df["TF-IDF"]
```

```python
def get_length(vector):
    s = 0
    for item in vector.values():
        s += item**2
    s = math.sqrt(s)
    return s
```

```python
def Normalize(vector):
    d = get_length(vector)
    for item in vector.keys():
        vector[item] = vector[item]/d
    return vector    
```

```python
df["TF-IDF"] = df["TF-IDF"].apply(Normalize)
```

```python
def get_word(discourNo, index):
    for key in df["TF-IDF"][discourNo].keys():
        if df["TF-IDF"][discourNo][key] == index:
            #print(key)
            break
    return key   
```

```python
def get_most(discourNo):
    D = {}
    ore = list(df["TF-IDF"][discourNo].values())
    ore.sort(reverse = True)
    ore = ore[:20]
    for elem in ore:
        D[get_word(discourNo, elem)] = elem
    return D
    
```

```python
def sum_values(liste, word):
    s = 0
    for tuplet in liste:
        if tuplet[0] == word:
            s += tuplet[1]
    return s        
            
```

```python
def one_for_all(df):
    L = []
    D = {}
    for elem in df["TF-IDF"].iteritems():
        L += list(elem[1].items())
    
    for tuplet in L:
        word = tuplet[0]
        if word not in D.keys():
            D[word] = sum_values(L, word)
            
    return D    
            
```

```python
D1 = one_for_all(df_d1)
```

```python
df_d2 = df.loc[df["year"]>2009]
```

```python
D2 = one_for_all(df_d2)
```

```python
def get_word_decade(dic, index):
    for key in dic.keys():
        if dic[key] == index:
            #print(key)
            break
    return key  
```

```python
def get_most_decade(dic):
    D = {}
    ore = list(dic.values())
    ore.sort(reverse = True)
    ore = ore[:20]
    for elem in ore:
        D[get_word_decade(dic, elem)] = elem
    return D
```
